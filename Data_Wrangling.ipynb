```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

filename = r'C:\Users\steph\bepec\Machine Learning\Stream Lit\automobile_pricing_predictor\data\Automobile price data _Raw_.csv'
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
auto_price = pd.read_csv(filename)
auto_price.columns = headers  #To name the columns in proper format
auto_price.head(5)
```

## Identifying missing values
 
### Convert ? to NAN

### In the car dataset, missing data comes with the question mark "?". We replace "?" with NaN (Not a Number), which is Python's default missing value marker for reasons of computational speed and convenience.

```python
auto_price.replace("?", np.nan, inplace=True)
auto_price.head(5)
```

## Evaluting for the missing data

### The missing values are converted to Python's default. We use Python's built-in functions to identify these missing values. There are two methods to detect missing data:

### .isnull()
### .notnull()
### The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.

```python
missing_data = auto_price.isnull()
missing_data.head(5)
```

## Count missing values in each column
### Using a for loop in Python, we can quickly figure out the number of missing values in each column."True" represents a missing value, "False" means the value is present in the dataset. In the body of the for loop the method ".value_counts()" counts the number of "True" values.

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

## Replace by mean:

### "normalized-losses": 41 missing data, replace them with mean
### "stroke": 4 missing data, replace them with mean
### "bore": 4 missing data, replace them with mean
### "horsepower": 2 missing data, replace them with mean

### "num-of-doors": 2 missing data, replace them with "four".
### Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur
### Drop the whole row:

## Data Standardization
### Data is usually collected from different agencies with different formats. (Data Standardization is also a term for a particular type of data normalization, where we subtract the mean and divide by the standard deviation)

```python
# The formula for unit conversion is L/100km = 235 / mpg
auto_price["city-L/100km"] = 235 / auto_price["city-mpg"]
```

## Data Normalization
### Why normalization?

### Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1

```python
auto_price["length"] = auto_price["length"]/auto_price["length"].max()
auto_price["width"] = auto_price["width"]/auto_price["width"].max()
auto_price["height"] = auto_price["height"]/auto_price["height"].max()
```

### Here we can see, we've normalized "length", "width" and "height" in the range of [0,1].

```python
auto_price[["length", "width", "height"]].head(5)
```

## Binning
### Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.

```python
auto_price["horsepower"]  = auto_price["horsepower"].astype(int, copy=True)  
```

### PLot the histogram to see what the distribution of horsepower looks like

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(auto_price["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("Horsepower bins")
```

### 3 bins of equal size bandwidth so we use numpy's linspace(start_value, end_value, numbers_generated function.

### Since we are building 3 bins of equal length, there should be 4 dividers, so numbers_generated=4.

```python
bins = np.linspace(min(auto_price['horsepower']), max(auto_price["horsepower"]), 4)
```

```python
bins
```

```python
#We set the group names
group_names = ["low","medium","high"]
```

### We apply the function "cut" the determine what each value of "df['horsepower']" belongs to.

```python
auto_price["horsepower-binned"] = pd.cut(auto_price["horsepower"], bins, labels=group_names, include_lowest=True)
```

```python
auto_price[["horsepower","horsepower-binned"]].head(5)
```

```python
#Lets see the number of vehicles in each bins
auto_price["horsepower-binned"].value_counts()
```

```python
# Lets plot the distribution of each bin

%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(group_names,auto_price["horsepower-binned"].value_counts())

plt.xlabel("horsepower")
plt.ylabel("counts")
plt.title("Horsepower plot")
```

### By checking the dataframe above carefully, we can find that the last column provides the bins for "horsepower" with 3 categories ("Low","Medium" and "High").

### We successfully narrow the intervals from 57 to 3!

```python
auto_price.head(5)
```

## Indicator variable (or dummy variable)
### An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning.

```python
auto_price.columns
```

```python
dummy_variable_1 = pd.get_dummies(auto_price["fuel-type"])
```

```python
dummy_variable_1.head(5)
```

```python
dummy_variable_1.rename({'fuel-type':'gas', 'fuel-type':'diesel'}, inplace=True)
```

```python
dummy_variable_1.head(5)
```

```python
auto_price = pd.concat([auto_price,dummy_variable_1], axis=1)
```
